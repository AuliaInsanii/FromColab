## **Decision Tree Regression**

Pohon keputusan adalah alat pendukung keputusan yang menggunakan model keputusan. Pohon keputusan biasanya digunakan research operasi atau pengambilan keputusan. Data yang sangat besar akan dibagi ke kelompok yang kecil sesuai dengan atrbutnya, dimana data dibagi terus menerus sampai data sesuai dengan kelompok. Ini mirip dengan manusia yang mana memecah data sesuai dengan pengalamannya sedangkan komputer memecah data sesuai dengan information gen.

Decision tree merupakan supervised learning yang menggantikan manusia membuat keputusan sesuai dengan data masa lalu. Decision tree digunakan untuk klasifikasi dan prosesnya dapat dsimulasikan sebagai pohon sehingga decision tree dapat memberitahukan bagaimana model dalam decison tree.

Cara untuk menggunakan Decision Tree

1. CART (Classification and Regression Tree)
Merupakan turunan dari metode lain yaitu CA45. CA45 merupakan turunan dari IDI Tree. Pada dasarnya IDI Tree dilakukan **splitting atau pemecahan data** berdasarkan data yang ada dalam dua kelompok dengan mengukur suatu angka yang disebut **entropi**. Entropi adalalh seberapa acak data itu. Semakin rendah entropi menunjukkan kelompok data semakin homogen atau sejenis atau kelompok yang sama. Jika entropi =  0 maka data sudah homogen semua atau seragam atau tidak acak. Jika entropi =  1 maka kelompok data sudah acak atau 50:50.
Misalnya, data sedan semua maka entropinya adalah 0. Jika saya memiliki 20 data sedan dan 20 data minibus maka data 50:50 dan nilai entropi = 1. Data dikelompokkan dari atas ke bawah sampai nilai **entropinya semakin rendah atau 0** atau data benar-benar sudah homogen. 

Untuk mencegah decision tree semakin panjang atau overfitting. Overfitting adalah modelnya bagus di datanya saja dan ketika digunakan pada data yang baru menjadi tidak bagus. Mencegah overfitting, dilakukan prunning atau pemangkasan pada pohon. Cabang yang terlalu panjang di node - node dan node yang paling atas dijadikan big node.

```python
import pandas as pd

df = pd.read_csv("/content/decisiontree_ch6.csv")
df
```

```python
encoding = {"mesin":{"bensin":0, "diesel":1}, "penggerak":{"depan":0, "belakang":1}}

df.replace(encoding,inplace=True)
```

```python
df
```

```python
X=df.drop(["ID", "label"], axis=1)
y=df["label"]
```

```python
import sklearn.model_selection as ms

X_train, X_test, y_train, y_test = ms.train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
import sklearn.tree as tree
# max_depth : maximun kedalam tree
#criterion : kriteria untuk decision tree
model = tree.DecisionTreeClassifier(criterion='entropy', max_depth=1) 
model.fit(X_train, y_train)

tree.DecisionTreeClassifier(criterion='entropy', max_depth=1)
```

```python
y_prediksi = model.predict(X_test)
y_prediksi
```

```python
import sklearn.metrics as met
print(met.accuracy_score(y_test, y_prediksi))
```

```python
import pydotplus as pp
labels = ['mesin', 'bangku', 'penggerak']
dot_data = tree.export_graphviz(model, out_file=None, feature_names=labels, filled=True, rounded=True)
graph = pp.graph_from_dot_data(dot_data)
graph.write_png('decisiontree.png')
```
